### Test `ClaudeAgent` with `biomedical-research` skill

- **pubmed-search**: Search PubMed/MEDLINE for biomedical and life sciences literature
- **arxiv-search**: Search arXiv for preprints in physics, math, CS, and quantitative biology

Both skills are automatically loaded from the `./skills` directory.

They can be used by the agent when relevant to the research task.

```python
import os
import anthropic
from open_claude_agent import ClaudeAgent

# Initialize Anthropic client
client = anthropic.Anthropic(api_key=os.environ.get("ANTHROPIC_API_KEY"))

# Create agent with biomedical skills automatically loaded
agent = ClaudeAgent(
    client=client,
    max_web_searches=10,  # Set reasonable limit for research tasks
    bash_timeout=120  # Increase timeout for network operations (PubMed/arXiv searches)
)

print("âœ“ ClaudeAgent created with biomedical research skills!")
print("âœ“ Available skills: pubmed-search, arxiv-search, web-research")
print("âœ“ Skills will be automatically discovered and used when relevant")
print("âœ“ Bash timeout set to 120 seconds for network operations")
```
